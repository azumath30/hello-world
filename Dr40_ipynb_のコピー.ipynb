```python
#
# * ２分探索のリスト表現
#

class Person:
    def __init__(self, left, name, right):
        self.left = left
        self.name = name
        self.right = right

nil = -1
a = [Person(  1, 'Machilda',  2),
     Person(  3, 'Candy'   ,  4),
     Person(  5, 'Rolla'   ,nil),
     Person(nil, 'Ann'     ,nil),
     Person(  6, 'Emy'     ,  7),
     Person(nil, 'Nancy'   ,nil),
     Person(nil, 'Eluza'   ,nil),
     Person(nil, 'Lisa'    ,nil)
    ]

keyname = 'Ann'
p = 0
while p != nil:
    if keyname == a[p].name:
        print('{:s}はみつかりました'.format(keyname))
        break
    elif keyname < a[p].name: # 左部分木へ移動
        p = a[p].left
    else:
        p = a[p].right
if p == nil:
    print('{:s}はみつかりませんでした'.format(keyname))
```
